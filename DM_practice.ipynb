# 1. Cluster Analysis 군집분석

군집분석은 지도학습과 비지도학습으로 나뉘는데, 보통 비지도학습으로 분류.
즉, 레이블이 주어지지 않으며 하지만 특정한 특징을 가지는 데이터셋.

```python
# estimated_value와의 관계를 알아보고자 함. 그러나 우리가 이미 관계를 알고 있어 지도학습이 될 수 있으니 삭제
X = df.drop('estimated_value', axis=1) #estimated_value를 제외한 모든 열
X = X[['bedrooms','bathrooms','rooms','squareFootage','lotSize','yearBuilt','priorSaleAmount']]
X.fillna(0, inplace=True) #Nan값을 갖고 있는 필드들을 모두 0으로 교체
```

```python
X_KMeans = KMeans(n_clusters=5, random_state=0).fit(X) #적합한 군집의 갯수를 찾기 위해서는 여러번 시도할 필요가 있음
```

random_state 매개변수는 KMeans 알고리즘에서 초기 중심(centroid)을 선택할 때 사용되는 난수 생성기의 시드를 설정합니다. K-means는 초기 중심의 위치에 따라 결과가 달라질 수 있기 때문에, 같은 데이터를 여러 번 학습시킬 때 결과가 일관되도록 하기 위해 random_state를 설정합니다.
...라고 하네요 by chat GPT

```python
print(X_KMeans.labels_) #각 데이터포인터 어느 클러스터에 포함되어 있는지 알 수 있음(0부터 4까지)
len(X_KMeans.labels_), X.shape #label의 갯수와 shape을 통해 확인할 수 있는 행의 갯수가 동일해야 함. 그래야 누락이 없다는 뜻이겠지?
```

```python
X_KMeans.cluster_centers_ #각 군집의 중심위치를 알려줌
X_KMeans.cluster_centers_[0]
```

```python
X['cluster'] = X_KMeans.labels_ #cluster열을 새롭게 만들어서 어떤 군집에 들어있는지 정보 추가
X.head(2)
```

```python
X.groupby('cluster').mean() #groupby로 묶음으로써 각 군집의 평균에 대한 정보를 알 수 있음
```

```python
from sklearn.metrics import silhouette_score #군집 내의 군집이 얼마나 잘 이뤄졌는지 보기 위해 라이브러리 추가
```

```python
#군집 분석을 여러분 반복하여 최적의 군집을 찾는 것의 예시 : for문의 이용
X =  X.drop('cluster', axis=1)
for i in range(3, 10) :
    newX_KMeans = KMeans(n_clusters=i).fit(X)
    labels = newX_KMeans.labels_
    print(silhouette_score(X, labels)) #이 부분의 값으로 군집 내의 군집이 얼마나 잘 이뤄졌는지 알 수 있음
    print(newX_KMeans.cluster_centers_)
```
